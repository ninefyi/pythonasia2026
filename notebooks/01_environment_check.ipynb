# 01 - Environment Check

Run every cell in order. Do not continue to the next notebook until all checks pass.

```python
import os
import sys
from pathlib import Path

print("Python:", sys.version)
assert sys.version_info >= (3, 10), "Python 3.10+ is required"
print("Python version check passed")
```

```python
required_packages = ["voyageai", "pymongo", "dotenv", "numpy"]
missing = []
for pkg in required_packages:
    try:
        __import__(pkg)
    except Exception:
        missing.append(pkg)

if missing:
    raise RuntimeError(f"Missing packages: {missing}. Run: uv pip install -r django_app/requirements.txt")
print("Package import checks passed")
```

```python
from dotenv import load_dotenv

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
load_dotenv(project_root / "django_app" / ".env")

MONGODB_URI = os.getenv("MONGODB_URI", "").strip()
voyage_api_key = os.getenv("VOYAGE_API_KEY", "").strip()

assert MONGODB_URI, "MONGODB_URI is not set in .env or django_app/.env"
assert voyage_api_key, "VOYAGE_API_KEY is not set in .env or django_app/.env"
print("Environment variables check passed")
```

```python
from pymongo import MongoClient
import voyageai

print("Testing MongoDB connection...")
print("MONGODB_URI:", MONGODB_URI)

client = MongoClient(MONGODB_URI, serverSelectionTimeoutMS=5000)
client.admin.command("ping")
print("MongoDB ping passed")

print("Testing Voyage API connection...")

vo = voyageai.Client(api_key=voyage_api_key)
emb = vo.embed(["hello workshop"], model="voyage-4-large").embeddings[0]
print("Voyage embedding check passed. Dimensions:", len(emb))
```
