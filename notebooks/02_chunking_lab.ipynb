# 02 - Chunking Lab

Goal: see how chunk size and overlap change what the model can understand.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

sample_plot = (
    "A team of explorers travels through a wormhole in space to find a new home for humanity. "
    "They face time dilation, hard choices, and the limits of love and science. "
    "The mission becomes a race to save people on Earth before it is too late."
)
print(sample_plot)
```

```python
def show_chunks(chunk_size, chunk_overlap):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
    )
    docs = splitter.create_documents([sample_plot])
    print(f"chunk_size={chunk_size}, chunk_overlap={chunk_overlap}, chunks={len(docs)}")
    for i, d in enumerate(docs, start=1):
        print(f"\nChunk {i}:")
        print(d.page_content)

show_chunks(chunk_size=80, chunk_overlap=0)
```

```python
show_chunks(chunk_size=80, chunk_overlap=20)
```
