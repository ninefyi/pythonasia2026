# 03 - Embeddings Lab

Goal: generate vectors from text and inspect dimensions and values.

```python
import os
from pathlib import Path
from dotenv import load_dotenv
import voyageai

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
load_dotenv(project_root / "django_app" / ".env")

api_key = os.getenv("VOYAGE_API_KEY", "").strip()
assert api_key, "Set VOYAGE_API_KEY in .env or django_app/.env"
vo = voyageai.Client(api_key=api_key)
```

```python
texts = [
    "A space adventure with explorers and a wormhole",
    "A family animation about toys coming to life",
    "A crime drama about a mafia family",
]
result = vo.embed(texts, model="voyage-4-large")
vectors = result.embeddings
print("Vectors:", len(vectors))
print("Dimensions:", len(vectors[0]))
print("First 5 values of vector 1:", [round(x, 6) for x in vectors[0][:5]])
```

```python
for text, vec in zip(texts, vectors):
    magnitude = sum(v * v for v in vec) ** 0.5
    print(f"Text: {text}")
    print(f"Vector magnitude: {magnitude:.4f}\n")
```
