# 04 - Similarity Lab

Goal: implement cosine similarity and rank results for a query.

```python
import math

def cosine_similarity(a, b):
    dot = sum(x * y for x, y in zip(a, b))
    norm_a = math.sqrt(sum(x * x for x in a))
    norm_b = math.sqrt(sum(y * y for y in b))
    return dot / (norm_a * norm_b) if norm_a and norm_b else 0.0
```

```python
query = [0.9, 0.1, 0.0]
docs = [
    ("space movie", [0.95, 0.05, 0.0]),
    ("crime movie", [0.2, 0.8, 0.1]),
    ("animation", [0.1, 0.2, 0.9]),
]

scored = []
for title, vec in docs:
    scored.append((title, cosine_similarity(query, vec)))

for title, score in sorted(scored, key=lambda x: x[1], reverse=True):
    print(f"{title}: {score:.4f}")
```
