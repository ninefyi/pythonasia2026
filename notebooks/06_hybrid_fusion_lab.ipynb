# 06 - Hybrid Fusion Lab

Goal: combine vector and keyword ranking with Reciprocal Rank Fusion (RRF).

```python
def rrf_fuse(vector_results, keyword_results, k=60):
    scores = {}

    for rank, doc in enumerate(vector_results):
        doc_id = doc["title"]
        scores.setdefault(doc_id, 0.0)
        scores[doc_id] += 1.0 / (k + rank + 1)

    for rank, doc in enumerate(keyword_results):
        doc_id = doc["title"]
        scores.setdefault(doc_id, 0.0)
        scores[doc_id] += 1.0 / (k + rank + 1)

    return sorted(scores.items(), key=lambda x: x[1], reverse=True)
```

```python
vector_results = [
    {"title": "Interstellar"},
    {"title": "The Matrix"},
    {"title": "Inception"},
]
keyword_results = [
    {"title": "The Matrix"},
    {"title": "Interstellar"},
    {"title": "The Avengers"},
]

fused = rrf_fuse(vector_results, keyword_results, k=60)
for title, score in fused:
    print(f"{title}: {score:.6f}")
```
